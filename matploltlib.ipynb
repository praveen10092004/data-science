```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['Lion','Horse','Kangaroo']
values1 = [15,24,30]
values2 = [20,18,25]

#setting the width of the bars
bar_width = 0.35
#caliculating bar positions for both groups
bar_positions1 = np.arange(len(categories))
bar_positions2 = bar_positions1 + bar_width


#creating the first set of bars (Group 1)
plt.bar(bar_positions1 , values1, width = bar_width, label='2021', color='skyblue')
#creating the second set of bars (Group 2) next to the first set
plt.bar(bar_positions2 , values2, width = bar_width, label='2022', color='orange')

#adding label to the axes
plt.xlabel('categories')
plt.ylabel('values')

#adding a title to the graph
plt.title('Grouped bar graph')

#displaying
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
#defining categories and values for two groups
categories = ['transport','incoe tax','housing']
values1 = [15,24,30]
values2 = [20,18,25]

#creating the first set of bars (group1) without any of
plt.bar(categories,values1, label='first half',color = 'skyblue')

#creating the second set of bars(group2) plotted without any offset
#this make group 2 bars stacked on top of group 1 bars
plt.bar(categories,values2,bottom = values1, label='second half',color='orange')

#adding labels to the axes
plt.xlabel('categories')
plt.ylabel('values')
plt.legend()
plt.title('stacked bar graph')
plt.show()
```

```python
#cumulative histogram
import matplotlib.pyplot as plt
import numpy as np

#generate random exam scores (out of 100)
exam_score = np.random.randint(0,100,150)
print(exam_score)

#create a cumulative histogram
plt.hist(exam_score, bins = 20,cumulative = True ,color='orange',edgecolor='white')
plt.xlabel('Exam Scores')
plt.ylabel('Cumulative No of Students')
plt.title('Cumulative Histogram of Exam Scores')
plt.show()
```

```python
#satcked histogram with multiple datasets
import matplotlib.pyplot as plt
import numpy as np

#sample data from two datasets
mark1 = np.array([2,4,5,7,9,10,11,13,14,15])
mark2 = np.array([6,7,8,10,11,12,13,14,15,16])

#creating a stacked histogram with different datasets
plt.hist([mark1,mark2],bins = 10,stacked = True,color=['skyblue','orange'],edgecolor='black')
plt.xlabel('Marks')
plt.ylabel('Count')
plt.title('satcked histogram with multiple datasets')
plt.legend(['Mark dataset 1','Mark dataset 2'])
plt.show()
```

```python
import pandas as pd
#series generated from list
a = [1,7,2]
myvar = pd.Series(a)
print(a)
print(myvar)
print("------------------------------")
#labels
print(myvar[0])
print("-----------------------------")
#assigning seperate labels to values in the series
myvar2 = pd.Series(a,index = ["x","y","z"])
print(myvar2)
print(myvar2["y"])
print("-----------------------------")
#series generated from dictionary
calories = {"day1":420,"day2":380,"day3":390}
myvar = pd.Series(calories)
print(myvar)
print("------------------------------")
myvar =pd.Series(calories,index = ["day1","day2"])
print(myvar)
#series generated from scale value
series = pd.Series(5, index =[0,1,2,3])
print(series)
print("-----------------------------")
series = pd.Series(5)
print(series)
```

```python
#Integer series
int_series = pd.Series([1,2,3,4])
print(int_series)
print("--------------------")
#Float series
float_series = pd.Series([1.0,2.0,3.0,4.0])
print(float_series)
print("--------------------")
#string series
str_series = pd.Series(['a','b','c','d'])
print(str_series)
print('--------------------')
```

```python
#labeled-based indexing
data = {'a':10,'b':20,'c':30}
series = pd.Series(data)
print(series['a'])
print(series[['a','c']])
print("------------------")
#Boolean indexing
data = [10,20,30,40,50]
series = pd.Series(data)
#Access the elements greater than 25
print(series[series > 25])
print("-------------------")
```

```python
#Dataframes
data = {
    'Name':['alice','Bob','Charlie'],
    'Age':[25,30,35],
    'city':['New york','Los Angeles','Chicago']
}
df = pd.DataFrame(data)
print(df)
print("----------------")
#selecting columns
print(df.Name)
print(df['Name'])
print(df[['Name','city']])
print("--------------------")
#select a single row by label
print(df.loc[0])
print(df.iloc[0])
#select multiple rows
print(df.iloc[0:2])
print("-------------------")

```

```python
#conditional selection
data = {
    'Name':['alice','Bob','Charlie'],
    'Age':[25,30,35],
    'city':['New york','Los Angeles','Chicago']
}
df = pd.DataFrame(data)
#select rows where age is greater then 25
print(df[df['Age']> 25])
print("----------------")
print(df.loc[(df.Name == 'Bob') & (df.city == 'Los Angeles')])
```
